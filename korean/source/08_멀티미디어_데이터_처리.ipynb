```python
import numpy as np

numpyArray2d = np.array([[1,1,1],[2,2,2],[3,3,3]])
print('2차원 배열 형상',numpyArray2d.shape, ', 데이터\n', numpyArray2d)
numpyArray3d = numpyArray2d[np.newaxis,:,:]
print('앞에 행이 추가된 3차원 배열 형상',numpyArray3d.shape, ', 데이터\n', numpyArray3d)
numpyArray3d = numpyArray2d[:,np.newaxis,:]
print('중간에 행이 추가된 3차원 배열 형상',numpyArray3d.shape, ', 데이터\n', numpyArray3d)
numpyArray3d = numpyArray2d[:,:,np.newaxis]
print('뒤에 행이 추가된 3차원 배열 형상',numpyArray3d.shape, ', 데이터\n', numpyArray3d)
```

```python
import numpy as np

numpyArray2d = np.array([[1,1,1],[2,2,2],[3,3,3]])
print('2차원 배열 형상',numpyArray2d.shape, ', 데이터\n', numpyArray2d)
numpyArray3d = np.expand_dims(numpyArray2d,axis=0)
print('0번 축으로 행이 추가된 3차원 배열 형상',numpyArray3d.shape, ', 데이터\n', numpyArray3d)
numpyArray3d = np.expand_dims(numpyArray2d,axis=1)
print('1번 축으로 행이 추가된 3차원 배열 형상',numpyArray3d.shape, ', 데이터\n', numpyArray3d)
numpyArray3d = np.expand_dims(numpyArray2d,axis=2)
print('2번 축으로 행이 추가된 3차원 배열 형상',numpyArray3d.shape, ', 데이터\n', numpyArray3d)
```

```python
import cv2
import numpy as np

videoCapture = cv2.VideoCapture('drive/MyDrive/videos/FilmModeDetection.mp4')

returnCode, frame = videoCapture.read()
frames = frame[np.newaxis,:,:,:]

while True:
    returnCode, frame = videoCapture.read()
    if not returnCode: break
    frame = frame[np.newaxis,:,:,:]
    frames = np.append(frames,frame,axis=0)

print('동영상의 형상:',frames.shape)
print('첫 프레임 이미지 우측 하단 잔디의 색상\n',frames[0][-1][-1:-6:-1])
```

```python
import cv2
import numpy as np

videoCapture = cv2.VideoCapture('drive/MyDrive/videos/FilmModeDetection.mp4')

count = int(videoCapture.get(cv2.CAP_PROP_FRAME_COUNT))
height = int(videoCapture.get(cv2.CAP_PROP_FRAME_HEIGHT))
width = int(videoCapture.get(cv2.CAP_PROP_FRAME_WIDTH))

frames = np.empty(shape=(count, height, width, 3))

for i in range(count):
    returnCode, frame = videoCapture.read()
    if not returnCode: break
    frames[i] = frame

print('동영상의 형상:',frames.shape)
print('첫 프레임 이미지 우측 하단 잔디의 색상\n',frames[0][-1][-1:-6:-1])
```

```python
videoCapture.release()
cv2.destroyAllWindows()
```
